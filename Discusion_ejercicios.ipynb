# **Ejercicio 4.5.2: Estados Separables y Entrelazados**  

En un sistema cuántico, las partículas pueden estar en estados separables o estados entrelazados, dependiendo de si pueden o no escribirse como producto de estados individuales de cada subsistema.  


## **Estados Separables**  
Un estado cuántico es separable si puede escribirse como el producto de los estados individuales de cada subsistema.  

### **Ejemplo con dos espines $ \frac{1}{2} $:**  
$
| \psi \rangle_{\text{total}} = | \uparrow \rangle_1 \otimes | \downarrow \rangle_2
$  
En este tipo de estado, medir la partícula 1 no afecta ni da información sobre la medición de la partícula 2.  


## **Estados Entrelazados**  
Un estado cuántico es entrelazado si no puede escribirse como el producto de estados individuales.  

### **Ejemplo:**  
$
\frac{1}{\sqrt{2}} \left( | \uparrow \rangle_1 \otimes | \downarrow \rangle_2 - | \downarrow \rangle_1 \otimes | \uparrow \rangle_2 \right)
$  
En este caso, las mediciones de los subsistemas están entrelazados, medir un subsistema afecta la descripción y el posible resultado del otro subsistema.  

## **Ejemplo: Entrelazamiento en Posición**  

Considerando el siguiente estado para dos partículas:  

$
| \psi \rangle = | x_0 \rangle \otimes | y_0 \rangle + | x_1 \rangle \otimes | y_1 \rangle
$

Para verificar si este estado es separable o entrelazado, podemos expresarlo como:  

$
| \psi \rangle = 1 \cdot | x_0 \rangle \otimes | y_0 \rangle + 0 \cdot | x_0 \rangle \otimes | y_1 \rangle + 0 \cdot | x_1 \rangle \otimes | y_0 \rangle + 1 \cdot | x_1 \rangle \otimes | y_1 \rangle
$

Si intentamos escribir $ | \psi \rangle $ como el producto tensorial de dos estados de los subsistemas, nos damos cuenta de que no es posible. En otras palabras, este estado no es separable, sino entrelazado.  





# **Ejercicio 4.5.3: Estado de Bell**  

Siguiendo con los estados separables y entrelazados, podemos observar el siguiente sistema:  

$
| \psi \rangle = | x_0 \rangle \otimes | y_0 \rangle + | x_1 \rangle \otimes | y_1 \rangle.
$

Para determinar si este estado es separable, debemos verificar si puede escribirse como el producto tensorial de dos estados individuales, es decir, si podemos encontrar dos estados  
$ | \phi_1 \rangle $ y $ | \phi_2 \rangle $ tales que:  

$
| \psi \rangle = | \phi_1 \rangle \otimes | \phi_2 \rangle.
$

Sin embargo, al intentar factorizar el estado, nos damos cuenta de que no es posible escribirlo en esta forma. Esto significa que el estado es entrelazado.  

Si analizamos el sistema en términos de espín, podemos definir una partícula con espín total cero. Al hacer esto, observamos una propiedad clave:  

- Si la primera partícula tiene espín $| \uparrow \rangle$, la segunda debe tener espín $ | \downarrow \rangle $.  
- Si la primera partícula tiene espín $ | \downarrow \rangle $, la segunda debe tener espín $| \uparrow \rangle $.  

Esto nos lleva al estado de Bell, un estado altamente entrelazado:  

$
| \psi \rangle = \frac{| \uparrow \rangle \otimes | \downarrow \rangle + | \downarrow \rangle \otimes | \uparrow \rangle}{\sqrt{2}}.
$
Este estado tiene una propiedad fundamental: si medimos el espín de la primera partícula, automáticamente sabemos el espín de la segunda.
